### Use beautifulsoup to pull the pokemon's name, number, and fighting type out of the elements in the table

```python
import urllib2
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(urllib2.urlopen('http://www.serebii.net/diamondpearl/pokemon.shtml').read())
soup.head()
```

```python
table = soup.find_all('table', width='98%')
table

raw ={
    'name': [],
    'number' : [],
    'type1' : [],
    'type2' : []
}

for idx, t in enumerate(table):
    for idx2, row in enumerate(t.find_all('tr')):
#         print '<=== row ===>', idx2
#         print row
        if (idx2 > 0):
            for idx3, col in enumerate(row.find_all('td')):
                if idx3 == 0:
                    raw['number'] += [col.text.replace('\r','').replace('\n','').replace('\t','').replace('#','')]
                if idx3 == len(row.find_all('td')) - 4:
                    raw['name'] += [col.text.replace('\n','')]
                if idx3 == len(row.find_all('td')) - 2:
                    key = col.find('a').contents
                    raw['type1'] += [key[0]['src'].replace('/pokedex-dp/type/','').replace('.gif','')]
                if idx3 == len(row.find_all('td')) - 1:
                    key = col.find('a').contents
                    raw['type2'] += [key[0]['src'].replace('/pokedex-dp/type/','').replace('.gif','')]
```

### Put the pokemon data into a dataframe and print it

```python
import pandas as pd
df = pd.DataFrame(raw)

# df['type2'] = df['type2'].replace('na','')
df['fighting type'] = df['type1'] + '-' + df['type2']
df['fighting type'] = [x.replace('-na','') for x in df['fighting type']]

df.drop(['type1','type2'], axis=1, inplace=True)
df
```
